```python
def APIData(x):
    s = "https://www.fema.gov/api/open/v1/" + x
    import requests 
    response = requests.get(s)
    data = response.json()
    df = pd.DataFrame.from_dict(data[x])
    return df

for i in apiList:
    #df = APIData(i)
    v = "df_" + i
    vars()[v] = APIData(i)
```
