```python
# Import Individual Assistance Housing Registrants Large Disasters - V1
import requests 
response = requests.get("https://www.fema.gov/api/open/v1/IndividualAssistanceHousingRegistrantsLargeDisasters")
# 200 means API was successful
print(response.status_code)
```
