## FEMA open data project

The purpose of this notebook is:
- to pull data from FEMA's open data website with an API
- to pull data from census.gov and other sources of economic indicators
- explore the relationship between individual assistance and public assistance against economic data in order to see if "certain areas recover faster or stronger than others"


"This could lead into a line of inquiry into what interventions, FEMA or otherwise, could have caused these better outcomes."

```python
import pandas as pd
apiList = ['IndividualAssistanceHousingRegistrantsLargeDisasters','HousingAssistanceOwners','HousingAssistanceRenters']
```

```python
def APIData(x):
    s = "https://www.fema.gov/api/open/v1/" + x
    import requests 
    response = requests.get(s)
    data = response.json()
    df = pd.DataFrame.from_dict(data[x])
    return df

for i in apiList:
    v = "df_" + i
    vars()[v] = APIData(i)
```
